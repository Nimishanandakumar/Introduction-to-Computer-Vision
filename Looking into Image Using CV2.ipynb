```python
img_bgr =cv2.imread('00-puppy.jpg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread("00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

```python
img_rgb.shape
```

```python
img = cv2.resize(img_rgb,(650,200))
plt.imshow(img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

```python
new_img =cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
new_img=cv2.flip(new_img,-1)
plt.imshow(new_img)
```

```python
type(new_img)
```

```python
cv2.imwrite('my_new_puppy.jpg',new_img)
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.imshow(new_img)
```
