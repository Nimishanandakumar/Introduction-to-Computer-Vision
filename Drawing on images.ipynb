## Drawing on Images

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

## Shapes

#### Rectangle

```python
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,128),color=(0,255,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```

#### Circle

```python
cv2.circle(blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```

#### Filled IN

```python
cv2.circle(blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

#### Lines

```python
cv2.line(blank_img,pt1=(0,0),pt2=(511,511),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

#### Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Shapes',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

#### Polygons

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```

### Adding Functionality with event choices

```python
import cv2
import numpy as np

def draw_circle(event,x,y,flags,param):
    if event ==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),-1)
        
        
img = np.zeros((512,512,3),np.uint8)
cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()        
```
